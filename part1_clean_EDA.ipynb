## Problem Statement

Kickstarter has recorded a 40% success rate for projects on its platform to date. Kickstarter earns a 5% commission as revenue for every successful project, which also means that for every unsuccessful project, Kickstarter is losing potential revenue due to its all-or-nothing funding model. They want to increase the success rate in order to maximise revenue.

### Stakeholders

1. Project creators: What information they can include to increase chances of success
2. Kickstarter: Identify higher risk projects to support
3. Investors: Identify projects likely to succeed for investment

### Business Objectives

Kickstarter as a business has a goal of maximising profit. This means it is best to have as many projects succeed as possible. Assuming an average of $2.2 million per year in missed potential profit from failed projects since 2009, by converting this could potentially lead to an increase in revenue of several million dollars per year.

“How can Kickstarter reduce the number of unsuccessful projects on the platform in order to increase revenue and grow the platform? What are the important factors behind success?”


### Project Objectives

We translate the business objectives into the following project objectives. 
Can we predict the outcome of a given project from project information available prior to launch?
Can we identify the key factors behind successful projects?

```python
# this function accesses the month subfolder containing csv files
# then combines data from all files into a single dataframe

def combine_csv_files_to_dataframe(month):
    
    # path to subfolder
    path = 'data/' + month

    # get all files in subfolder in name order 
    files = sorted(os.listdir(path))

    # list of dataframes for all data within a given month
    dfs = []
    for file in files:
        filepath = path + '/' + file
        
        # read single csv
        df = pd.read_csv(filepath)
        
        # create new column 'snapshot_date' and convert to date type
        # this will be used to remove duplicate records later
        df['snapshot_date'] = month
        df['snapshot_date'] = pd.to_datetime(df['snapshot_date'], format='%Y_%m')
        
        # add to list
        dfs.append(df)
    
    # combine into single dataframe
    dfs = pd.concat(dfs, axis=0)
    
    return dfs
            
```

### Text 

Natural Language Processing (NLP) is used to analyse the text columns such as project name and blurb.

```python
import re
```

```python
def clean_text(text):
     
    # add space after special characters excluding '’ for contraction - note the different apostrophes??
    # some have no space after full stop e.g. excellent.Good service 
    # this ensures the words are separated
    text = re.sub(r'[^a-zA-Z\s\']', ' ', text)
    
    # remove extra space
    text = re.sub(r'\s+', ' ', text)
    
    return text
```

```python
# compute polarity and subjectivity
# Polarity (-1 to 1), subjectivity (0 to 1)

def get_sentiment_scores(text):
    text_blob = textblob.TextBlob(text)
    return text_blob.sentiment.polarity, text_blob.sentiment.subjectivity
```

```python
# example of text columns

df[['slug', 'name', 'blurb']].head()
```

NLP will not be applied to 'slug' as it contains redundant information to project name.

```python
# clean project names
df['name'] = df['name'].apply(clean_text)

# get project name word count
df['name_word_count'] = df['name'].apply(lambda x: len(x.split(" ")))

# compute name polarity and subjectivity, rounded to 2dp
df[['name_polarity', 'name_subjectivity']] = df['name'].apply(get_sentiment_scores).apply(pd.Series).round(2)
```

A few records are missing blurb. We will impute these with the text from 'slub'.

```python
df[df['blurb'].isnull()]
```

```python
# impute with words from slug column, replacing - with space
df['blurb'] = df['blurb'].fillna(df['slug'].str.replace('-', ' '))
```

```python
# clean project blurb
df['blurb'] = df['blurb'].apply(clean_text)

# get project blurb word count
df['blurb_word_count'] = df['blurb'].apply(lambda x: len(x.split(" ")))

# compute name polarity and subjectivity, rounded to 2dp
df[['blurb_polarity', 'blurb_subjectivity']] = df['blurb'].apply(get_sentiment_scores).apply(pd.Series).round(2)
```

### Category

How many projects have there been under each category and what is the success rate?

```python
def plot_categorical_data(data, feature):
    fig, (ax1, ax2, ax3) =  plt.subplots(3, 1, figsize=(12,14))
    
    # project count by feature
    ax1 = sns.countplot(data=data, x=feature, ax=ax1, order=data[feature].value_counts().index, palette='rainbow')
    ax1.set(title=f'Number of projects by {feature}')
    ax1.set_xlabel('')
    
    # project count by feature and status
    ax2 = sns.countplot(data=data, x=feature, hue='status', palette=palette, ax=ax2, 
                        order=data[feature].value_counts().index)
    ax2.set(title=f'Number of projects by {feature} and status')
    ax2.set_xlabel('')
    
    # total amount pledged by feature
    pledged_by_feature = data.groupby(feature)['usd_pledged'].sum().reset_index(name='total_usd_pledged')
    pledged_by_feature['total_usd_pledged'] = pledged_by_feature['total_usd_pledged'] / 1000000
    
    ax3 = sns.barplot(data=pledged_by_feature, y='total_usd_pledged', x=feature, ax=ax3,
                      order=data[feature].value_counts().index, palette='rainbow')
    ax3.set(title=f'Total pledged (USD) by {feature} since 2009')
    ax3.set_ylabel('USD (Millions of $)')
    
    # common plot settings
    for ax in [ax1, ax2, ax3]:
        ax.set_frame_on(False)
        ax.set_xticklabels(ax1.get_xticklabels(), rotation=90)
    
    # some padding between each row
    plt.subplots_adjust(hspace=0.5)
    plt.show()
    
```

```python
plot_categorical_data(df, 'category')
```

Top categories are film & video, music and art. Most projects have been based in US. 

Technology has the highest total pledged since 2009, with 'games' and 'film & video' being next highest. Dance, crafts and journalism has the least amount pledged. Categories that have higher proportion of failed projects are technology, food, crafts, photography and journalism.

### Distribution by Goal & Pledged


```python
# this helper function takes in 2 features and plots boxplot, violinplot and histogram side by side.

def plot_compare_features_by_status(data, feature1, feature2, title1, title2, y1, y2):
    
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(10,12))

    sns.boxplot(y=feature1, x='status', data=data, palette=palette, ax=ax1, fliersize=2)
    ax1.set_title(f'Boxplot - {title1}')
    ax1.set_xlabel('')
    ax1.set_ylabel(y1)

    sns.boxplot(y=feature2, x='status', data=data, palette=palette, ax=ax2, fliersize=2)
    ax2.set_title(f'Boxplot - {title2}')
    ax2.set_xlabel('')
    ax2.set_ylabel(y2)

    sns.violinplot(y=feature1, x='status', data=data, palette=palette, ax=ax3)
    ax3.set_title(f'Violinplot - {title1}')
    ax3.set_xlabel('')
    ax3.set_ylabel(y1)

    sns.violinplot(y=feature2, x='status', data=data, palette=palette, ax=ax4)
    ax4.set_title(f'Violinplot - {title2}')
    ax4.set_xlabel('')
    ax4.set_ylabel(y2)

    sns.histplot(data=data,
                 x=feature1,
                 bins=50,
                 hue='status',
                 hue_order=hist_hue_order,
                 palette=palette,
                 ax=ax5)
    ax5.set_title(f'{title1} by Status')
    ax5.set_xlabel(y1)

    sns.histplot(data=data,
                 x=feature2,
                 bins=50,
                 hue='status',
                 hue_order=hist_hue_order,
                 palette=palette,
                 ax=ax6)
    ax6.set_title(f'{title2} by Status')
    ax6.set_xlabel(y2)

    # some padding between each row
    plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

We will scale the goal and pledged columns with log transformation as the numbers can get very large. Presence of outliers makes it difficult to see the real trend.

```python
# e.g. without scaling
plt.figure(figsize=(6, 3))
sns.boxplot(y='usd_goal', x='status', data=df, palette=palette, fliersize=2)
plt.show()
```
